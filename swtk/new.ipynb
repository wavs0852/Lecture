```python
import random

def extractor(time, lst_t):
    lst = []
    
    while len(lst) < time:
        value = random.choice(lst_t)
        if value not in lst:
            lst.append(value)
        else:
            value = random.choice(lst_t)
        
    lst.sort()
    return lst

for i in range(100):
    nnew = ['P' + str(i + 1) + s for i in range(6) for s in ['A', 'B']]
    nnnw = nnew * 2
    a = extractor(4, nnnw)
    print(a)
```
