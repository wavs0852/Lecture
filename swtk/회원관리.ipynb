```python
class Get():
    def __init__(self, id, name=''):
        self.id = id; self.name = name

class Manage():
    def __init__(self):
        self.members = {}

    def member_add(self, member):
        self.members[member.id] = member.name
        print(f"ID: {member.id}, 이름: {member.name}을 추가했습니다.")

    def member_remove(self, id):
        if not len(self.members) == 0:
            if self.members[id]:
                del self.members[id]
                print(f"ID: {id}의 회원정보를 삭제했습니다.")
            else:
                print(f"ID: {id}인 회원정보는 존재하지 않습니다.")
        else:
            print('회원이 존재하지 않습니다.')

    def member_print_one(self, id):
        if id in self.members:
            print("=== 회원정보 ===")
            print(f"ID: {id}, 이름: {self.members[id]}")
        else:
            print(f"ID: {id}인 회원정보는 존재하지 않습니다.")

    def member_print_all(self):
        if not len(self.members) == 0:
            print("=== 전체회원 ===")
            for id in self.members.keys():
                print(f"ID: {id}, 이름: {self.members[id]}")
        else:
            print('회원이 존재하지 않습니다.')

default = Manage()

while True:
    print(f'+---------+---+\n| 회원출력 | 1 |\n| 폴더출력 | 2 |\n| 회원추가 | 3 |\n| 기능정지 | 4 |\n+---------+---+\n')
    usr = int(input())
    if usr == 1:
        default.member_print_one(input('ID를 입력해주세요.'))
    elif usr == 2:
        default.member_print_all()
    elif usr == 3:
        number, person = input('ID를 입력해주세요.'), input('이름을 입력해주세요.')
        p1 = Get(number, person) 
        default.member_add(p1)
    elif usr == 4:
        break
    else:
        print('잘못된 입력입니다.')
```
