##### 문제 최적화

```python
import random
import pandas as pd

# 주어진 자원
lst_lec = ['P' + str(i + 1) + s for i in range(6) for s in ['A', 'B']]
lst_prf = ['확률과통계/고선우', '딥러닝/고선우', 'AI알고리즘/권수태', '기계학습/권수태', 'AI수학/민정익', '논리적문제해결/민정익', '인공지능기초/이근호', '서비스러닝/이근호', '논리적문제해결/송주환', '영상이해/송주환', 'AI프로그래밍/김영수', '리눅스운영체제/김영수']
dct_lec = dict(zip(lst_lec, lst_prf))|{'JOB':'진로탐색', 'BOB':'점심시간'}

# 중복검색 알고리즘
def search(lst):
    dummy, flag = '000', False
    for value in lst:
        if value[1] == dummy[1]:
            flag = True
        dummy = value
    return flag

# 배열 알고리즘
while True:
    lst_lecture = lst_lec * 2
    choose = random.sample(lst_lecture, len(lst_lecture))
    if ((len(set(choose[0:10])) == 10) and (len(set(choose[5:14])) == 9) and (len(set(choose[10:19])) == 9) and (len(set(choose[14:24])) == 10) and (len(set([i[1] for i in choose[0:5]])) >= 4) and (len(set([i[1] for i in choose[5:10]])) >= 4) and (len(set([i[1] for i in choose[10:14]])) >= 3) and (len(set([i[1] for i in choose[14:19]])) >= 4) and (len(set([i[1] for i in choose[19:24]])) >= 4)):
        chosen = [choose[0:5], choose[5:10], choose[10:14], choose[14:19], choose[19:24]]
        for i in range(len(chosen)):
            while search(chosen[i]) == True:
                chosen[i] = random.sample(chosen[i], len(chosen[i]))
            chosen[i].insert(2, 'BOB')
        chosen[2].append('JOB')
        break

# 판다스 변환
for y in range(len(chosen)):
    for x in range(len(chosen[y])):
        chosen[y][x] = dct_lec[chosen[y][x]]

df = pd.DataFrame(chosen, index=['월','화','수','목','금'], columns=['09:00','10:30','12:00','01:00','02:30','04:00'])
df.T
```

```python
import numpy as np
def JeonjuAIClass():
    global week
    global karma
    AIclass=['KO1','KO2','KW1','KW2','MN1','MN2','LE1','LE2','SO1','SO2','KM1','KM2']
    week=np.zeros([6,5])
    week=week.astype(str)
    week[2][:]='XXX'        #점심시간 제외
    week[5][2]='XXX'        #진로탐색 제외
    for c in AIclass:
        karma=0
        selected=0
        while selected!=2:          #한 과목당 2개씩 고르기
            for i in range(week.shape[0]):
                if selected==2:
                    break
                for j in range(week.shape[1]):
                    candy='ignore'
                    if selected==2:
                        break
                    if karma>100:
                        return 'x'
                    lucky=np.random.randint(1,5)
                    if lucky==1:
                        if j==1 or j==2 or j==3:
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j+1]) and (c not in week[:,j-1]): #j행이 양 끝이 아닐 때(화요일,수요일,목요일). 양 끝에 있을 때 j+1과 j-1을 사용하면 꼬이기 때문.  
                                candy=c
                        elif j==0:        #왼쪽일 때(월요일)
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j+1]):
                                candy=c
                        elif j==4:          #오른쪽일 때(금요일)
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j-1]):
                                candy=c
                        if candy!='ignore':          #여기부터는 시간 간격 벌리기용
                            if i==0:  #9시일때(위 없음)
                                if candy[0:2] not in week[i+1][j]:
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                            elif i==1 or i==2 or i==3 or i==4:
                                if (candy[0:2] not in week[i+1][j]) and (candy[0:2] not in week[i-1][j]):
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                            elif i==5:  #2시반일때(밑 없음)
                                if candy[0:2] not in week[i-1][j]:
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                        else:
                            karma+=1
    return 'o'
while True:
    a=JeonjuAIClass()
    if a=='o':
        break
print(week)
```

### 고한웅 풀이

```python
import random

# 요일별 리스트 생성
mon = []
tue = []
wed = []
thu = []
fri = []

# 숫자와 문자 대응 딕셔너리
number_to_letter = {
    1: "확통",
    2: "딥러닝",
    3: "ai알고",
    4: "기계학습",
    5: "ai수학",
    6: "논문1",
    7: "ai기초",
    8: "서비스",
    9: "논문2",
    10: "영상이해",
    11: "ai코딩",
    12: "리눅스",
   
}

# 1부터 12까지의 숫자를 랜덤으로 선택하여 요일별 리스트에 추가 (연속된 요일 간에 겹치지 않게)
while len(mon) < 5:
    random_number = random.randint(1, 12)
    if random_number not in mon:
        mon.append(random_number)

while len(tue) < 5:
    random_number = random.randint(1, 12)
    if random_number not in mon and random_number not in tue:
        tue.append(random_number)



while len(wed) < 4:
    random_number = random.randint(1, 12)
    if random_number not in tue and random_number not in wed:
        wed.append(random_number)

while len(thu) < 5:
    random_number = random.randint(1, 12)
    if random_number not in wed and random_number not in thu:
        thu.append(random_number)

while len(fri) < 5:
    random_number = random.randint(1, 12)
    if random_number not in thu and random_number not in fri:
        fri.append(random_number)

# 결과 출력
print('         ','1교시','  ', ' 2교시','  ', ' 3교시','  ', ' 4교시','  ', ' 5교시')
print("월요일:", [number_to_letter[num] for num in mon])
print("화요일:", [number_to_letter[num] for num in tue])
print("수요일:", [number_to_letter[num] for num in wed]+["진로"])
print("목요일:", [number_to_letter[num] for num in thu])
print("금요일:", [number_to_letter[num] for num in fri])
```
