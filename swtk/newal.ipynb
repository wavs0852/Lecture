```python
import random
quo = 4
lst_table = [[],[],[],[],[]]

# 원소 추출 알고리즘
def extractor(time, lst_1):
    lst_local = []
    flag_lst = False

    for i in range(time): 
        chosen = random.choice(lst_1)
        if chosen in lst_local:
            flag_lst = True
        while flag_lst == True:
            index_1 = [v for v in lst_1 if not v == chosen]
            chosen = random.choice(index_1)
            if chosen not in lst_local:
                flag_lst = False

        lst_lecture.remove(chosen)
        lst_local.append(chosen)
        
    return sorted(lst_local)

# 배열 조합 알고리즘
def mkarray():
    lst_2 = []
    def isgood(enterlst):
        good_1 = list(set(enterlst[0]) & set(enterlst[2]))
        good_2 = list(set(enterlst[1]) & set(enterlst[3]))
        return len(good_1) + len(good_2)
    def repeat(want):
        dum = 'Dummy'
        for c in lst_2[i-1]:
            index_2 = [v for v in lst_lecture if not v == c]
            itsc = list(set(dum) & set(index_2))
            if dum == 'Dummy':
                dum = index_2 
            else:
                dum = itsc
        extracted = extractor(want, itsc) # quo만 함수 입력값으로 받기
        lst_2.append(extracted)

    for i in range(len(lst_table)):
        if i == 0:
            extracted = extractor(quo + 1, lst_lecture)
            lst_2.append(extracted)
        elif i == 1:
            repeat(quo+1)
        else:
            repeat(quo)

    # 나머지 분류 알고리즘
    fst, snd = lst_2[0] + lst_2[1], lst_2[2]
    fstnum, sndnum = 0, 0

    for i in lst_lecture:
        if i in fst:
            fstnum +=1
            fstvalue = i
        if i in snd:
            sndnum +=1
            sndvalue = i
    if fstnum == 1 and sndnum == 1:
        lst_2[4].append(sndvalue)
        lst_2[3].append(fstvalue)
        lst_2.append(isgood(lst_2))
        return lst_2
    elif fstnum == 2:
        for i in range(len(lst_lecture)):
            lst_2[i+3].append(lst_lecture[i])
        lst_2.append(isgood(lst_2))
        return lst_2
    else:
        return False

ziphab = []
for i in range(1000):
    lst_lecture = ['P' + str(i + 1) + s for i in range(6) for s in ['A', 'B']] * 2
    ziphab.append(mkarray())

# 우성 시간표 분류기
exam_1 = [t for t in ziphab if not t == False]
exam_2 = [g for g in exam_1 if g[5] > 6]
exam_2
```

```python
import numpy as np
def JeonjuAIClass():
    global week
    global karma
    AIclass=['KO1','KO2','KW1','KW2','MN1','MN2','LE1','LE2','SO1','SO2','KM1','KM2']
    week=np.zeros([6,5])
    week=week.astype(str)
    week[2][:]='X'        #점심시간 제외
    week[5][2]='X'        #진로탐색 제외
    for c in AIclass:
        karma=0
        selected=0
        while selected!=2:          #한 과목당 2개씩 고르기
            for i in range(week.shape[0]):
                if selected==2:
                    break
                for j in range(week.shape[1]):
                    candy='ignore'
                    if selected==2:
                        break
                    if karma>100:
                        return 'x'
                    lucky=np.random.randint(1,5)
                    if lucky==1:
                        if j==1 or j==2 or j==3:
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j+1]) and (c not in week[:,j-1]): #j행이 양 끝이 아닐 때(화요일,수요일,목요일). 양 끝에 있을 때 j+1과 j-1을 사용하면 꼬이기 때문.  
                                candy=c
                        elif j==0:        #왼쪽일 때(월요일)
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j+1]):
                                candy=c
                        elif j==4:          #오른쪽일 때(금요일)
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j-1]):
                                candy=c
                        if candy!='ignore':          #여기부터는 시간 간격 벌리기용
                            if i==0:  #9시일때(위 없음)
                                if candy[0:2] not in week[i+1][j]:
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                            elif i==1 or i==2 or i==3 or i==4:
                                if (candy[0:2] not in week[i+1][j]) and (candy[0:2] not in week[i-1][j]):
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                            elif i==5:  #2시반일때(밑 없음)
                                if candy[0:2] not in week[i-1][j]:
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                        else:
                            karma+=1
    return 'o'
while True:
    a=JeonjuAIClass()
    if a=='o':
        break
print(week)
```
