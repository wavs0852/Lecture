```python
import turtle
import random

turtle.setup(width = 1000, height = 1000)

t=turtle.Pen()

for i in range(10):
    t.up()
    t.goto(random.randint(-300,300), random.randint(-300,300))
    t.down()
    t.speed(0)
    len = random.randint(60,100)
    for x in range(1,19):
        t.forward(len)
        if x % 2 == 0:
            t.left(175)
        else:
            t.left(225)
```
