```python
import random
import pandas as pd

# 주어진 자원
lst_lec = ['P' + str(i + 1) + s for i in range(6) for s in ['A', 'B']]
lst_prf = ['확률과통계/고선우', '딥러닝/고선우', 'AI알고리즘/권수태', '기계학습/권수태', 'AI수학/민정익', '논리적문제해결/민정익', '인공지능기초/이근호', '서비스러닝/이근호', '논리적문제해결/송주환', '영상이해/송주환', 'AI프로그래밍/김영수', '리눅스운영체제/김영수']
dct_lec = dict(zip(lst_lec, lst_prf))|{'JOB':'진로탐색', 'BOB':'점심시간'}

# 중복검색 알고리즘
def search(lst):
    dummy, flag = '000', False
    for value in lst:
        if value[1] == dummy[1]:
            flag = True
        dummy = value
    return flag

# 배열 알고리즘
while True:
    lst_lecture = lst_lec * 2
    choose = random.sample(lst_lecture, len(lst_lecture))
    if ((len(set(choose[0:10])) == 10) and (len(set(choose[5:14])) == 9) and (len(set(choose[10:19])) == 9) and (len(set(choose[14:24])) == 10) and (len(set([i[1] for i in choose[0:5]])) >= 4) and (len(set([i[1] for i in choose[5:10]])) >= 4) and (len(set([i[1] for i in choose[10:14]])) >= 3) and (len(set([i[1] for i in choose[14:19]])) >= 4) and (len(set([i[1] for i in choose[19:24]])) >= 4)):
        chosen = [choose[0:5], choose[5:10], choose[10:14], choose[14:19], choose[19:24]]
        for i in range(len(chosen)):
            while search(chosen[i]) == True:
                chosen[i] = random.sample(chosen[i], len(chosen[i]))
            chosen[i].insert(2, 'BOB')
        chosen[2].append('JOB')
        break

# 판다스 변환
for y in range(len(chosen)):
    for x in range(len(chosen[y])):
        chosen[y][x] = dct_lec[chosen[y][x]]

df = pd.DataFrame(chosen, index=['월','화','수','목','금'], columns=['09:00','10:30','12:00','01:00','02:30','04:00'])
df.T
```

```python
import random
quo = 4
lst_table = [[],[],[],[],[]]

# 원소 추출 알고리즘
def extractor(time, lst_1):
    lst_local = []
    flag_lst = False

    for i in range(time): 
        chosen = random.choice(lst_1)
        if chosen in lst_local:
            flag_lst = True
        while flag_lst == True:
            index_1 = [v for v in lst_1 if not v == chosen]
            chosen = random.choice(index_1)
            if chosen not in lst_local:
                flag_lst = False

        lst_lecture.remove(chosen)
        lst_local.append(chosen)
        
    return sorted(lst_local)

# 배열 조합 알고리즘
def mkarray():
    lst_2 = []
    def isgood(enterlst):
        good_1 = list(set(enterlst[0]) & set(enterlst[2]))
        good_2 = list(set(enterlst[1]) & set(enterlst[3]))
        return len(good_1) + len(good_2)
    def repeat(want):
        dum = 'Dummy'
        for c in lst_2[i-1]:
            index_2 = [v for v in lst_lecture if not v == c]
            itsc = list(set(dum) & set(index_2))
            if dum == 'Dummy':
                dum = index_2 
            else:
                dum = itsc
        extracted = extractor(want, itsc) # quo만 함수 입력값으로 받기
        lst_2.append(extracted)

    for i in range(len(lst_table)):
        if i == 0:
            extracted = extractor(quo + 1, lst_lecture)
            lst_2.append(extracted)
        elif i == 1:
            repeat(quo+1)
        else:
            repeat(quo)

    # 나머지 분류 알고리즘
    fst, snd = lst_2[0] + lst_2[1], lst_2[2]
    fstnum, sndnum = 0, 0

    for i in lst_lecture:
        if i in fst:
            fstnum +=1
            fstvalue = i
        if i in snd:
            sndnum +=1
            sndvalue = i
    if fstnum == 1 and sndnum == 1:
        lst_2[4].append(sndvalue)
        lst_2[3].append(fstvalue)
        lst_2.append(isgood(lst_2))
        return lst_2
    elif fstnum == 2:
        for i in range(len(lst_lecture)):
            lst_2[i+3].append(lst_lecture[i])
        lst_2.append(isgood(lst_2))
        return lst_2
    else:
        return False

ziphab = []
for i in range(1000):
    lst_lecture = ['P' + str(i + 1) + s for i in range(6) for s in ['A', 'B']] * 2
    ziphab.append(mkarray())

# 우성 시간표 분류기
exam_1 = [t for t in ziphab if not t == False]
exam_2 = [g for g in exam_1 if g[5] > 6]
exam_2
```

```python
import numpy as np
def JeonjuAIClass():
    global week
    global karma
    AIclass=['KO1','KO2','KW1','KW2','MN1','MN2','LE1','LE2','SO1','SO2','KM1','KM2']
    week=np.zeros([6,5])
    week=week.astype(str)
    week[2][:]='XXX'        #점심시간 제외
    week[5][2]='XXX'        #진로탐색 제외
    for c in AIclass:
        karma=0
        selected=0
        while selected!=2:          #한 과목당 2개씩 고르기
            for i in range(week.shape[0]):
                if selected==2:
                    break
                for j in range(week.shape[1]):
                    candy='ignore'
                    if selected==2:
                        break
                    if karma>100:
                        return 'x'
                    lucky=np.random.randint(1,5)
                    if lucky==1:
                        if j==1 or j==2 or j==3:
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j+1]) and (c not in week[:,j-1]): #j행이 양 끝이 아닐 때(화요일,수요일,목요일). 양 끝에 있을 때 j+1과 j-1을 사용하면 꼬이기 때문.  
                                candy=c
                        elif j==0:        #왼쪽일 때(월요일)
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j+1]):
                                candy=c
                        elif j==4:          #오른쪽일 때(금요일)
                            if (week[i][j]==str('0.0')) and (c not in week[:,j]) and (c not in week[:,j-1]):
                                candy=c
                        if candy!='ignore':          #여기부터는 시간 간격 벌리기용
                            if i==0:  #9시일때(위 없음)
                                if candy[0:2] not in week[i+1][j]:
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                            elif i==1 or i==2 or i==3 or i==4:
                                if (candy[0:2] not in week[i+1][j]) and (candy[0:2] not in week[i-1][j]):
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                            elif i==5:  #2시반일때(밑 없음)
                                if candy[0:2] not in week[i-1][j]:
                                    week[i,j]=c
                                    selected+=1
                                else:
                                    karma+=1
                        else:
                            karma+=1
    return 'o'
while True:
    a=JeonjuAIClass()
    if a=='o':
        break
print(week)
```
