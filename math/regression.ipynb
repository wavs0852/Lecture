### 인수분해

```python
from sympy import expand, factor, Symbol
x = Symbol('x')

one = expand((x + 1) * (x + 5)) # 수식전개
two = factor(one) # 인수분해

one,two
```

### 지수함수, 로그함수

```python
import math

pow = math.pow(2, 5) # 2의 5제곱
sqrt = math.sqrt(2) # 2의 제곱근
exp = math.exp(2) # 자연상수(e)의 2제곱

print(f'pow = {pow}\nsqrt = {sqrt}\nexp = {exp}')
```

### 지수함수, 로그함수의 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) 

plt.plot(x, y)
plt.show()
```

### 시그모이드 함수

numpy로 풀기

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
y = sigmoid(x)

print(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1) # x값
y = sigmoid(x) # y값

plt.plot(x, y)
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

x = [-1.0, 1.0, 2.0]
y = [sigmoid(i) for i in x]

print(y)
```

### 함수의 극한

x가 무한대로 갈 때의 극한

```python
from sympy import Limit, S, Symbol

x = Symbol('x')
Limit(1/x, x, S.Infinity).doit()
```

우극한

```python
from sympy import Limit, S, Symbol

x = Symbol('x')

Limit(1/x, x, 0).doit()
```

좌극한

```python
from sympy import Limit, S, Symbol

x = Symbol('x')
Limit(1/x, x, 0, dir='-').doit()
```

수렴

```python
# 수렴 판별기
from sympy import Limit, S, Symbol

def calc(sic, a):
    if Limit(sic, x, a).doit() == Limit(sic, x, a, dir='-').doit():
        return True
    else:
        return False
    
x = Symbol('x')

calc(1/x, 2)
```

연속

```python
# 연속 판별기
from sympy import Limit, S, Symbol

def calc(equation, a):
    if Limit(equation, x, a).doit() == Limit(equation, x, a, dir='-').doit() == equation.subs(x, a):
        return True
    else:
        return False

x = Symbol('x')
calc(1/x, 0)
```
